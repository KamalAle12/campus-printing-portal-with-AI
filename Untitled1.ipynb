```python
#Q-Learning Algorithm
import numpy as np
import random

#parameters
num_states = 5 #Example number of states(queue lengths)
num_actions = 2 #Example number of actions(e.g., serve or delay)
num_episodes = 100

alpha = 0.1  # Learning rate
gamma = 0.9  # Discount factor
epsilon = 0.1  # Exploration rate

# Initialize Q-Table
Q = np.zeros((num_states, num_actions))

def choose_action(state):
    if random.uniform(0, 1)<epsilon:
        #Explore: choose a random action
        return random.randint(0, num_actions-1)
    else:
        # Exploit: choose the best action based on Q-values
        return np.argmax(Q[state, :])

def get_reward(state, action):
    # Define the reward function for the system
    # Example reward: higher reward for serving the queue faster
    if action == 0: #serve
        return 1 if state > 0 else -1  # Reward if the queue is served
    else:  # Delay
        return -1  # Penalty for delaying
def update_q_table(state, action, reward, next_state):
    best_next_action = np.argmax(Q[next_state, :])
    Q[state, action] += alpha * (reward + gamma * Q[next_state, best_next_action] - Q[state, action])

# Training
for episode in range(num_episodes):
    state = random.randint(0, num_states - 1)  # Initial state
    done = False
    while not done:
        action = choose_action(state)
        reward = get_reward(state, action)
        next_state = state + (1 if action == 0 else -1)  # Example state transition

        # Ensure next_state is within bounds
        next_state = max(0, min(num_states - 1, next_state))
        update_q_table(state, action, reward, next_state)
        
        state = next_state
        if state == 0:  # Example condition for ending the episode
            done = True

# Print final Q-Table
print("Final Q-Table:")
print(Q)
```

```python
#optimizing resource allocation, Linear Programming (LP) 
from scipy.optimize import linprog

# Objective function coefficients (profit coefficients, for maximization)
c = [-10, -20]  # Coefficients need to be negative for minimization in linprog

# Coefficients for inequality constraints (Ax <= b)
A = [
    [2, 1],  # Coefficients for the first constraint
    [1, 2]   # Coefficients for the second constraint
]
b = [20, 30]  # Right-hand side values for the constraints

# Bounds for each variable (e.g., non-negative)
x_bounds = (0, None)
bounds = [x_bounds, x_bounds]

# Solve the problem
result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Display the results
if result.success:
    print('Optimal value:', -result.fun)  # Negate to get the original objective value
    print('Optimal solution:', result.x)
else:
    print('No optimal solution found.')
```

```python
#Visualization
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog

# Define the objective function coefficients (for maximization, hence negated)
c = [-10, -20]

# Define the constraints
A = [
    [2, 1],  # Coefficients for the first constraint
    [1, 2]   # Coefficients for the second constraint
]
b = [20, 30]

# Bounds for variables
x_bounds = (0, None)
bounds = [x_bounds, x_bounds]

# Solve the LP problem
result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Extract the results
optimal_value = -result.fun
optimal_solution = result.x

# Define the constraint lines
x = np.linspace(0, 15, 400)
y1 = (20 - 2 * x)  # from 2x1 + x2 <= 20
y2 = (30 - x) / 2  # from x1 + 2x2 <= 30

# Plot constraints
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=r'$2x_1 + x_2 \leq 20$')
plt.plot(x, y2, label=r'$x_1 + 2x_2 \leq 30$')

# Fill feasible region
plt.fill_between(x, np.minimum(y1, y2), where=(x >= 0) & (np.minimum(y1, y2) >= 0), color='gray', alpha=0.5)

# Plot optimal solution
plt.scatter(optimal_solution[0], optimal_solution[1], color='red', label='Optimal Solution', zorder=5)
plt.annotate(f'Optimal Solution\n({optimal_solution[0]:.2f}, {optimal_solution[1]:.2f})',
             xy=(optimal_solution[0], optimal_solution[1]),
             xytext=(optimal_solution[0]+2, optimal_solution[1]-2),
             arrowprops=dict(facecolor='red', shrink=0.05))

# Plot settings
plt.xlim(0, 15)
plt.ylim(0, 15)
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.title('Linear Programming: Resource Allocation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Parameters
num_records = 1000
np.random.seed(42)

# Generate synthetic data
data = pd.DataFrame({
    'job_size': np.random.randint(1, 50, num_records),  # Job size
    'cluster': np.random.randint(0, 3, num_records),  # Cluster (e.g., based on K-Means clustering)
    'user_type': np.random.choice(['student', 'staff'], num_records)  # User type
})

# Convert categorical variable to numerical
data['user_type_num'] = data['user_type'].map({'student': 0, 'staff': 1})

# Features and target
X = data[['job_size', 'cluster']]
y = data['user_type_num']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize and train the model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Make predictions
y_pred = rf.predict(X_test)
```

```python
# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['student', 'staff'], yticklabels=['student', 'staff'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
